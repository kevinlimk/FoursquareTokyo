```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
df_X0 = pd.read_csv('df_X0.csv',
                    index_col = 0,
                    parse_dates = [0],
                    date_parser = pd.to_datetime,
                    infer_datetime_format = True)
```

```python
df_X1 = pd.read_csv('df_X1.csv',
                    index_col = 0,
                    parse_dates = [0],
                    date_parser = pd.to_datetime,
                    infer_datetime_format = True)
```

```python
# Load predictions from Layer 0
df_layer0_results = pd.read_csv("layer0_results.csv")
```

```python
ytest0 = df_layer0_results.ytest0.values
ypred0 = df_layer0_results.ypred0.values
```

```python
calc_accuracy_ypred0 = list(ypred0[ypred0 < 2])
calc_accuracy_ytest0 = list(ytest0[np.where(ypred0 < 2)])
```

```python
df_X1.head()
```

```python
df_X1.tail()
```

```python
Xtrain1 = df_X1['2012-July':'2013-Jan'].drop(["y1"], axis = 1)
ytrain1 = df_X1['2012-July':'2013-Jan'].y1

Xtest1 = df_X1['2013-Feb' : ].drop(["y1"], axis = 1)
ytest1 = df_X1['2013-Feb' : ].y1

clf_LR1 = LogisticRegression()
clf_LR1.fit(Xtrain1, ytrain1)
ypred1 = clf_LR1.predict(Xtest1)
```

```python
print("Accuracy: {0:.2f}%".format(100 * accuracy_score(list(calc_accuracy_ytest0) + list(ytest1), list(calc_accuracy_ypred0) + list(ypred1))))
```

```python
print("Micro-F1 score: {0:.2f}".format(f1_score(list(calc_accuracy_ytest0) + list(ytest1), list(calc_accuracy_ypred0) + list(ypred1), average = "micro")))
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(Xtrain1, ytrain1)
ypred_knn = knn.predict(Xtest1)
```

```python
print("Accuracy: {0:.2f}%".format(100 * accuracy_score(list(calc_accuracy_ytest0) + list(ytest1), list(calc_accuracy_ypred0) + list(ypred_knn))))
```

```python
print("Micro-F1 score: {0:.2f}".format(f1_score(list(calc_accuracy_ytest0) + list(ytest1), list(calc_accuracy_ypred0) + list(ypred_knn), average = "micro")))
```
